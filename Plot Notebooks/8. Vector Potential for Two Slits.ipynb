```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import emproperties as emp

from scipy.constants import constants as sc
from scipy.special import ellipk
```

```python
%matplotlib inline
%config InlineBackend.figure_format='retina'
sns.set_context("poster")
sns.set_style("ticks",{"xtick.direction" : u"in",
                       "ytick.direction" : u"in"})
sns.set_palette(sns.color_palette("coolwarm", 4))
```

```python
a = 4
b = 4
d = 50e-09 
pen_depth=33.3e-09

em = emp.EMProperties(a=a, b=b, d=d, pen_depth=pen_depth)

# x = np.arange(-4,4,0.0005)
A = 1
Phi = ((sc.mu_0 *A)/em.b) * ellipk(em.kp)

x = np.arange(-10,10,0.005)
y = np.arange(-10,10,0.005)

X,Y = np.meshgrid(x,y)
Az = (-Phi/np.pi) * ( np.arctan((X-a)/abs(Y)) - np.arctan((X+a)/abs(Y)) )

# For normalisation
maxs = [max(Az[i]) for i in range(len(Az))]
```

```python
plt.figure(figsize=(12,8))

cmap='RdGy'
# cmap='winter'
plt.contourf(X/a,Y/a,Az/max(maxs),30,cmap=cmap)
cbar = plt.colorbar()
plt.xlabel('$x/a$')
plt.ylabel('$y/a$')
cbar.set_label('$A_{z}(x,y)$', rotation=270,labelpad=35)
plt.tight_layout()
plt.savefig('../figs/vector_potential_two_slits.pdf',bbox_inches=None)
plt.show()
```

```python
xlist = np.arange(-12,12,0.001)

a = 2
b = 10

em.a = a
em.b = b

y = 1

Kzx = []
for x in xlist:
    Phi = ((sc.mu_0 *A)/em.b) * ellipk(em.kp)
    Az0 = (-Phi/np.pi) * ( np.arctan((x-a)/abs(y)) - np.arctan((x+a)/abs(y)) )
    if abs(x)<=0:
        Kzx.append((2/sc.mu_0*em.pearl_length) * (Phi-Az0))
    if abs(x)>0: 
        Kzx.append((-2/sc.mu_0*em.pearl_length) * Az0)
```

```python
plt.plot(xlist/a,Kzx)
plt.show()
```
