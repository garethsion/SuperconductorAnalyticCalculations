```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import emproperties as emp

from scipy.constants import constants as sc
from scipy.special import ellipk
```

```python
%matplotlib inline
%config InlineBackend.figure_format='retina'
sns.set_context("poster")
sns.set_style("ticks",{"xtick.direction" : u"in",
                       "ytick.direction" : u"in"})
sns.set_palette(sns.color_palette("coolwarm", 4))
```

```python
em = emp.EMProperties(a=1, b=2, d=50e-09, pen_depth=33.3e-09)

x = np.arange(-4,4,0.0005)
A = 1
Phi = ((sc.mu_0 *A)/em.b) * ellipk(em.kp)
K = (-4*Phi*em.a )/(np.pi*sc.mu_0*x**2)

# Kz = em.current_density(xlist)
# Hy = em.complex_field(xlist)
```

```python
plt.plot(x/em.a,K*em.a)
plt.show()
```
