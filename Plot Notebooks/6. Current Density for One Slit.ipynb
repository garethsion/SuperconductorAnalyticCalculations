```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import emproperties as emp
```

```python
%matplotlib inline
%config InlineBackend.figure_format='retina'
sns.set_context("poster")
sns.set_style("ticks",{"xtick.direction" : u"in",
                       "ytick.direction" : u"in"})
sns.set_palette(sns.color_palette("coolwarm", 4))
```

```python
em = emp.EMProperties(a=1e-06, b=2e-06, d=50e-09, pen_depth=33.3e-09)

lset = 1 # set length top one to get the cross section

x = np.arange(-.5e-06,.5e-06,0.01e-06)

current_density_one_slit = em.current_density_one_slit(x)
```

```python
fig = plt.figure(figsize=(10,6))
ax0 = fig.add_subplot(111)
ax0.plot(x/em.a,np.array(current_density_one_slit)*em.a)

plt.text(.12,.4,'$a = 1 \mu m$',horizontalalignment='center',
         verticalalignment='center', transform=ax0.transAxes)
plt.text(.12,.3,'$b = 2 \mu m$',horizontalalignment='center',
         verticalalignment='center', transform=ax0.transAxes)
plt.text(.15,.2,'$d = 0.05 \mu m$',horizontalalignment='center',
         verticalalignment='center', transform=ax0.transAxes)

ax0.set_xlabel('$x/a$')
ax0.set_ylabel('$K_{0}(x)$')
ax0.set_title('Current-Density for One long Slit')
plt.savefig('figs/current_density_one_slit.pdf',bbox_inches=None)
plt.show()
```
